```python
#Make function to convert html rows to leg sets
def leggify(x):
    gmst = []
    for i in x:
        if (i.text == '') or (i.text == '\xa0'):
            gmst.append(0)
        else:
            gmst.append(i.get_text())
    return gmst
```

```python
leggify(rows[2])
```

```python
#Create a set of leg sets
all_legs = []
for row in rows[2:]:
    rowset = row.find_all('td')
    pleg = [leggify(rowset)]
    all_legs.append(pleg)
```

```python
def PName(matches):
    for leg in matches:
        if leg[0][0] == 0:
            print(leg[0][1])
        else:
            print(leg[0][2])
```

```python
def LegCode(matches):
    for leg in matches:
        if leg[0][0] != 0:
            thisleg = leg[0][0]
            print(leg[0][0])
        else:
            print(thisleg)
```

```python
def Game(matches):
    for leg in matches:
        if leg[0][0] != 0:
            thisgame = leg[0][1]
            print(leg[0][1])
        else:
            print(thisgame)    
```

```python
def P3DA(matches):
    for leg in matches:
        if leg[0][0] != 0:
            print(leg[0][5])
        elif (leg[0][0] == 0) & (leg[0][4] == 0):
            print(leg[0][2])
        else:
            print(leg[0][4])
```
